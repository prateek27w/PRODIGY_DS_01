## Top 10 countries with highest population in year 2022

```python
data_top=data_new.sort_values(by='2022',ascending=False)
data_set= data_top.head(10)
data_set
```

## Data Visualisation

```python
def add_value_labels(ax,spacing=30):
    for rect in ax.patches:
        x=rect.get_x()+rect.get_width()/2
        y=rect.get_height()-3
        
        va='bottom' if y>=0 else 'top'
        label= '()'.format(y) 
        y_shift = spacing * (1 if y >= 0 else -1)
        ax.annotate(label,(x,y),xytext=(0,y_shift),textcoords="offset points",ha='center',va=va)
```

## Creating Bar Graph with decade wise comparison of population for respective years


```python
plt.figure(figsize=(20,13))
plt.subplot(2,2,1)
plot=sns.barplot(x='1960',y="Country Name",data=data_set,palette="rocket")
plt.title("Top 10 countries Of Total Population in 1960",fontweight='bold')
plt.xlabel("Total population")
plt.ylabel("Countries")
add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.barplot(x='1970',y="Country Name",data=data_set,palette="rocket")
plt.title("Top 10 countries Of Total Population in 1970",fontweight='bold')
plt.xlabel("Total population")
plt.ylabel("Countries")
add_value_labels(plot)

plt.subplot(2,2,3)
plot=sns.barplot(x='1980',y="Country Name",data=data_set,palette="rocket")
plt.title("Top 10 countries Of Total Population in 1980",fontweight='bold')
plt.xlabel("Total population")
plt.ylabel("Countries")
add_value_labels(plot)

plt.subplot(2,2,4)
plot=sns.barplot(x='1990',y="Country Name",data=data_set,palette="rocket")
plt.title("Top 10 countries Of Total Population in 1990",fontweight='bold')
plt.xlabel("Total population")
plt.ylabel("Countries")
add_value_labels(plot)

plt.tight_layout()
```
